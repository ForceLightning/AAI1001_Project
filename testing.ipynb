```python
def perturb_to_mean(batched_inputs, batched_cam, step_size=0.25):
    mean = batched_inputs.mean(dim=1)
    perturbed_signal = (batched_inputs - mean.repeat(360, 1).T).numpy()
    diff = batched_cam * perturbed_signal
    diff *= step_size
    perturbed_signal = perturbed_signal + mean.repeat(360, 1).T.numpy() - diff
    return perturbed_signal
```

```python
explainability_set = ECGDataset("./data/heartbeats_evensplit_test/", target_transform=label_encode, item_transform=hb_transform)
perturbed_tensor = None
model_outputs = []
BATCH_SIZE = 2048
for i in tqdm(range(10), position=0, leave=True):
    # 10 iterations of perturbations
    loader = DataLoader(explainability_set, batch_size=BATCH_SIZE, shuffle=False, pin_memory=True, num_workers=8)
    ys, preds, outputs = [], [], []
    with GradCAM1D(model=model, target_layer=target_layers, use_cuda=True) as cam:
        for i, (in_tensor, target_tensor) in enumerate(tqdm(loader, position=1, leave=False)):
            target_category = target_tensor
            greyscale_cam = cam(in_tensor, target_tensor)
            perturbed_data = perturb_to_mean(in_tensor.detach(), greyscale_cam, step_size=0.1)
            end_idx = min((i+1)*BATCH_SIZE, len(explainability_set.X))
            explainability_set.X[i*BATCH_SIZE:end_idx] = perturbed_data
            with torch.no_grad():
                output = model(in_tensor.cuda()).cpu().detach()
                proba = torch.nn.functional.softmax(output, dim=1)
                pred = output.argmax(dim=1)
                ys.append(target_tensor.numpy())
                preds.append(pred.numpy())
                outputs.append(proba.numpy())
            torch.cuda.empty_cache()
        ys = np.concatenate(ys)
        preds = np.concatenate(preds)
        outputs = np.concatenate(outputs)
        model_outputs.append({
            "y": np.array(ys),
            "preds": np.array(preds),
            "outputs": np.array(outputs)
        })
        # loader._iterator._shutdown_workers()
```
